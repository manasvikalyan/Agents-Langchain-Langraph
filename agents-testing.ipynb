```python
from langchain_groq import ChatGroq
import os

groq_api_key = os.getenv("GROQ_API_KEY")

llm = ChatGroq(
    model_name="llama-3.1-8b-instant",
    temperature=0,
    max_tokens=None,
    max_retries=2,
)
```

```python
from langchain_core.tools import tool

@tool
def add_numbers(a: float, b: float) -> float:
    """Add two numbers and return the sum."""
    return a + b

@tool(return_direct=True)
def translate_text(text: str) -> str:
    """Translate text from English to French."""
    # assumes `llm` (ChatGroq) is already defined in a previous cell
    resp = llm.invoke(f"Translate this text to French. Only output the translation:\n\n{text}")
    # ChatGroq returns an AIMessage with `.content`
    return resp.content if hasattr(resp, "content") else str(resp)

tools = [add_numbers, translate_text]
```

```python
from langchain.agents import create_agent

agent = create_agent(
    model=llm,
    tools=tools,
    system_prompt="You are a helpful assistant",
)

# Run the agent once
result = agent.invoke({
    # "messages": [
    #     {"role": "user", "content": "3, 5"}
    # ]
    "messages": [
        {"role": "user", "content": "Convert 'how are you?' to French?"}
    ]
})

# The last message is the model's reply

# print the thinking and which tool was used


print(result["messages"][-1].content)

```
